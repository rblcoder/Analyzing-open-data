```python
import pandas as pd
import numpy as np
%matplotlib inline
import cufflinks as cf
cf.go_offline()
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_retail = pd.read_excel('./Online Retail.xlsx')
```

```python
df_retail['Price*Quantity'] = df_retail['UnitPrice']*df_retail['Quantity']
```

```python
pd.pivot_table(df_retail, index='Description', values='Price*Quantity', 
               aggfunc=np.sum).sort_values('Price*Quantity',
               ascending=False)[:10].iplot(kind='bar', logy=True)
```

```python
df_retail['Description'].value_counts()[:10].iplot(kind='bar', logy=True)
```

```python
df_retail.nunique()
```

```python
#https://markhneedham.com/blog/2015/01/25/python-find-the-highest-value-in-a-group/
df_retail.groupby("Country").max()["Price*Quantity"]
```

```python
df_retail.loc[df_retail.groupby("Country").idxmax()["Price*Quantity"]]
```

```python
pd.pivot_table(df_retail, index='InvoiceNo', values='Price*Quantity', 
               aggfunc=np.sum).sort_values('Price*Quantity',
               ascending=False)[:10]
```

```python
df_retail.loc[df_retail['InvoiceNo'].isin(pd.pivot_table(df_retail, index='InvoiceNo', values='Price*Quantity', 
               aggfunc=np.sum).sort_values('Price*Quantity',
               ascending=False)[:10].index)].sort_values(by='Price*Quantity', ascending=False)
```

```python
df_retail.groupby("InvoiceNo").max()["Price*Quantity"].reset_index().sort_values(by='Price*Quantity', ascending=False)[:10]
```
